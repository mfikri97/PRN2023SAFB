## logging in!

```python
def login_to_facebook(email, password, driver_path):
    # Specify Chrome options to disable unnecessary features
    options = Options()
    options.add_experimental_option("detach", True)
    options.add_argument("--disable-infobars")
    options.add_argument("--disable-extensions")
    options.add_argument("--disable-notifications")
    options.add_argument("--disable-popup-blocking")

    # Create the WebDriver instance
    service = Service(executable_path=driver_path)
    driver = webdriver.Chrome(service=service, options=options)
    # Navigate to the Facebook login page
    driver.get("https://www.facebook.com/login/")

    # Target the username and password input fields
    username_input = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='email']")))
    password_input = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='pass']")))

    # Enter the provided username and password
    username_input.clear()
    username_input.send_keys(email)
    password_input.clear()
    password_input.send_keys(password)

    # Target the login button and click it
    login_button = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']")))
    login_button.click()

    # Wait for the login to complete
    WebDriverWait(driver, 10).until(EC.title_contains("Facebook"))

    return driver

```

## SCROLLING

```python
def scroll_until_date_found(driver, post_link_class, stopping_date,personal_brand_url, scroll_attempts_limit=15):
    # Add a 2-second wait
    driver.get(personal_brand_url)
    time.sleep(2)

    # Scroll down the page until a post with a date of July is found
    SCROLL_PAUSE_TIME = 1
    scroll_attempts = 0
    stopping_date_found = False

    while scroll_attempts < scroll_attempts_limit and not stopping_date_found:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)
    
        # Check if the elements with the specific class and date of July exist
        date_elements = driver.find_elements("xpath", f"//a[@class='{post_link_class}']")
            
        for element in date_elements:
            if stopping_date in element.text:
                print("Found a post with a date of July!")
                # Process the element(s) as needed
                print("Post Text:", element.text)
                stopping_date_found = True  # Exit the loop if a post with a date of July is found
            
        scroll_attempts += 1
```

## SCRAPE MAIN POST

```python
def scrape_main_text(sub_text_class_1_post,sub_text_class_2_post, sub_text_class_1_video, sub_text_class_2_video):
    try:
        #For normal post
        sub_divs = driver.find_elements(By.XPATH, f".//div[contains(@class, '{sub_text_class_1_post}') or contains(@class, '{sub_text_class_2_post}')]")
    except:
        # For videos
        sub_divs = driver.find_elements(By.XPATH, f".//div[contains(@class, '{sub_text_class_1_video}') or contains(@class, '{sub_text_class_2_video}')]")
    complete_text = ""
    for sub_div in sub_divs:
        text = sub_div.text.strip()
        complete_text += text + "\n"

    return complete_text
    
```

## SCRAPE THE DATE

```python
def scrape_date(post_link_class):
    date_element = driver.find_elements(By.XPATH, f"//a[@class='{post_link_class}']")
    for item in date_element:
        date = item.text
        if 'Today' in date:
            date_posted = datetime.now()
        elif 'Yesterday' in date:
            days_ago = 1
            date_posted = datetime.now() - timedelta(days=days_ago)
        elif 'w' in date:
            weeks_ago = int(date.split()[0])
            date_posted = datetime.now() - timedelta(weeks=weeks_ago)  
        elif 'd' in date:
            days_ago = int(date.split()[0])
            date_posted = datetime.now() - timedelta(days=days_ago)                
        elif 'h' in date:
            date_posted = datetime.now()
        else:
            try:
                if 'at' in date:  # Handles format like "13 August at 22:09"
                    day_month_time = date.split(' at ')
                    day_month = day_month_time[0]
                    time = day_month_time[1]
                    year = datetime.now().year  # Take current year
                    formatted_date = f"{day_month} {year} at {time}"
                    date_posted = datetime.strptime(formatted_date, '%d %B %Y at %H:%M')
                elif len(date.split()) == 2:  # Handles format like "1 January"
                    day, month = date.split()
                    year = datetime.now().year  # Take current year
                    formatted_date = f"{day} {month} {year}"
                    date_posted = datetime.strptime(formatted_date, '%d %B %Y')
                elif len(date.split()) == 3:  # Handles format like "25 July 2023"
                    day, month, year = date.split()
                    formatted_date = f"{day} {month} {year}"
                    date_posted = datetime.strptime(formatted_date, '%d %B %Y')
                else:
                    date_posted = datetime.strptime(date, '%d %B %Y')
            except ValueError:
                date_posted = None
        
        if date_posted:
            date_posted = date_posted.strftime('%Y-%m-%d %H:%M:%S')
        break
    return date_posted
```

## SHARE AND COMMENT COUNT

```python
def share_comment_scraping(driver, comment_share_box_class, comment_class, share_class):
    try:
        sharencomment_box_div = driver.find_element(By.XPATH,f"//div[@class='{comment_share_box_class}']")
        # Find the div element containing the number of comments
        comment_div = sharencomment_box_div.find_elements(By.XPATH,f".//span[@class='{comment_class}']")
        # Find the div element containing the number of shares
        share_div = sharencomment_box_div.find_elements(By.XPATH,f".//span[@class='{share_class}'][1]")  # This index [1] assumes the share number is the second div with this class in the main_div, adjust if needed
        # Extract the number of comments and shares from the div elements
        # If comment_div is not found, set number_of_comments to "N/A"
        number_of_comments = comment_div[0].text if comment_div else "N/A"
        # If share_div is not found or the second element is not found, set number_of_shares to "N/A"
        number_of_shares = share_div[1].text if share_div and len(share_div) >= 2 else "N/A"

    except NoSuchElementException:
        # If any of the elements is not found, set them to None or other appropriate value
        sharencomment_box_div = None
        comment_div = None
        share_div = None
        number_of_comments = "N/A"
        number_of_shares = "N/A"
    return number_of_comments, number_of_shares
    
```

## REACTION SCRAPING

```python
def convert_numeric_value(value):
    if value.endswith('K'):
        return int(float(value[:-1]) * 1000)
    elif value.endswith('M'):
        return int(float(value[:-1]) * 1000000)
    else:
        return int(value)
```

```python
def reaction_scraping(reaction_element_class):
    # Find all elements containing the reactions and their corresponding counts
    reaction_elements = driver.find_elements(By.XPATH, f"//div[@class='{reaction_element_class}']")
    reactions_count = {}
    for reaction_element in reaction_elements:
        try:
            # Re-find the reaction element to avoid StaleElementReferenceException
            aria_label = reaction_element.get_attribute("aria-label")
            if aria_label is not None:
                try:
                    reaction_type, count_text = aria_label.split(":")
                    count_text = count_text.strip("people")
                    count_text = count_text.strip()
                    count = convert_numeric_value(count_text)
                    reactions_count[reaction_type] = count
                except ValueError:
                    continue
        except StaleElementReferenceException:
            continue
    return reactions_count
```

```python
def comment_scraping(post_url, comments_elements_class, comment_author_class, comment_text_class, author_class, comments_date_section_element_parent, comments_date_section_element_child):
    comment_elements = driver.find_elements(By.XPATH, f"//div[@class='{comments_elements_class}']")
    comment_data = []

    for comment_element in comment_elements:
        try:
            author_elements = driver.find_elements(By.XPATH, f"//a[@class='{author_class}']")
            if author_elements:
                author_name = author_elements[0].text.strip()
            else:
                author_name = "N/A"
            author_comment_element = comment_element.find_element(By.XPATH, f".//span[@class='{comment_author_class}']/span[@dir='auto']")
            author_comment = author_comment_element.text.strip()
            comment_text_element = comment_element.find_element(By.XPATH, f".//div[@class='{comment_text_class}']")
            comment_text = comment_text_element.text.strip()

            # Extract comment date
            date_elements = comment_element.find_elements(By.XPATH, f".//a[@class='{comments_date_section_element_parent}']")
            date_commented = None  # Initialize with None in case date not found
            for date_element in date_elements:
                comment_date_section = date_element.find_elements(By.XPATH, f".//span[contains(@class, '{comments_date_section_element_child}')]")
                if comment_date_section:
                    date = comment_date_section[0].text.strip()

                if 'Today' in date:
                    date_commented = datetime.now()
                elif 'Yesterday' in date:
                    days_ago = 1
                    date_commented = datetime.now() - timedelta(days=days_ago)
                elif 'd' in date:
                    days_ago = int(date.split()[0])
                    date_commented = datetime.now() - timedelta(days=days_ago)
                elif 'w' in date:
                    weeks_ago = int(date.split()[0])
                    date_commented = datetime.now() - timedelta(weeks=weeks_ago)
                elif 'h' in date:
                    date_commented = datetime.now()
                else:
                    try:
                        date_commented = datetime.strptime(date, '%d %B %Y')
                    except ValueError:
                        date_commented = None
                    
                if date_commented:
                    date_commented = date_commented.strftime('%Y-%m-%d %H:%M:%S')
                break  # Found date, exit the loop

            comment_data.append({"author": author_comment, "text": comment_text, "post_url": post_url, "host_author": author_name, "comment_date": date_commented})
        except:
            continue

    return comment_data
```

```python
def comment_scraping(post_url, comments_elements_class_firstcom, comments_elements_class_therest, comment_author_class, comment_text_class_firstline, comment_text_class_therest, author_class, comments_date_section_element_parent, comments_date_section_element_child):
    # Find all comment elements
    comment_elements_classes = f"div[@class='{comments_elements_class_firstcom}' or @class='{comments_elements_class_therest}']"
    comment_elements = driver.find_elements(By.XPATH, f"//{comment_elements_classes}")
    comment_data = []

    for comment_element in comment_elements:
        try:
            author_elements = driver.find_elements(By.XPATH, f"//a[@class='{author_class}']")
            if author_elements:
                author_name = author_elements[0].text.strip()
            else:
                author_name = "N/A"
            author_comment_element = comment_element.find_element(By.XPATH, f".//span[@class='{comment_author_class}']/span[@dir='auto']")
            author_comment = author_comment_element.text.strip()


            comment_text_elements_firstline = comment_element.find_elements(By.XPATH, f".//div[@class='{comment_text_class_firstline}']")
            comment_text_elements_secondline = comment_element.find_elements(By.XPATH, f".//div[@class='{comment_text_class_therest}']")
            comment_text_firstline = '\n'.join([element.text.strip() for element in comment_text_elements_firstline])
            comment_text_secondline = '\n'.join([element.text.strip() for element in comment_text_elements_secondline])
            comment_text = f"{comment_text_firstline}\n{comment_text_secondline}"


            # Extract comment date
            date_elements = comment_element.find_elements(By.XPATH, f".//a[@class='{comments_date_section_element_parent}']")
            date_commented = None  # Initialize with None in case date not found
            for date_element in date_elements:
                comment_date_section = date_element.find_elements(By.XPATH, f".//span[contains(@class, '{comments_date_section_element_child}')]")
                if comment_date_section:
                    date = comment_date_section[0].text.strip()

                if 'Today' in date:
                    date_commented = datetime.now()
                elif 'Yesterday' in date:
                    days_ago = 1
                    date_commented = datetime.now() - timedelta(days=days_ago)
                elif 'd' in date:
                    days_ago = int(date.split()[0])
                    date_commented = datetime.now() - timedelta(days=days_ago)
                elif 'w' in date:
                    weeks_ago = int(date.split()[0])
                    date_commented = datetime.now() - timedelta(weeks=weeks_ago)
                elif 'h' in date:
                    date_commented = datetime.now()
                else:
                    try:
                        date_commented = datetime.strptime(date, '%d %B %Y')
                    except ValueError:
                        date_commented = None
                    
                if date_commented:
                    date_commented = date_commented.strftime('%Y-%m-%d %H:%M:%S')
                break  # Found date, exit the loop

            comment_data.append({"author": author_comment, "text": comment_text, "post_url": post_url, "host_author": author_name, "comment_date": date_commented})
        except Exception as e:
            print("Exception:", e)
            traceback.print_exc()  # Print detailed traceback
            print("fail")


    return comment_data
```
